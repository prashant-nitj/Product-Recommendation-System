```python
import numpy as np
import pandas as pd
```

```python
# Load the laptops dataset
df = pd.read_csv('C:/Users/Desktop/laptops.csv')
```

```python
df
```

```python
df.columns
```

```python
df.info()
```

```python
to_drop1=['Cache', 'Mic In',
       'Speakers', 'Multi Card Slot', 'RJ45 (LAN)', 'Dedicated Graphics.1',
       'HDMI Port', 'Ethernet', 'Battery Life (up to hours)',
       'Dedicated Graphic Memory Type', 'Expandable RAM up to (GB)',
       'other_info','Internal Mic','Base Clock Speed'
    
]
df.drop(to_drop1,inplace=True,axis=1)
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df = df.fillna(method = 'ffill')
```

```python
df.isnull().sum()
```

```python
df= df.fillna(method = 'bfill')
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates()
```

```python
df.info()
```

```python
# Split data into training and testing sets
train_size = int(0.8 * len(df))
train_data = df[:train_size]
test_data = df[train_size:]
```

```python
# create a deep copy of test_data
test_data1 = test_data.copy(deep=True)
```

```python
# Get list of object columns
obj_cols = train_data.select_dtypes(include=['object']).columns.tolist()

# Loop through object columns and factorize
for col in obj_cols:
    train_data[col] = pd.factorize(train_data[col])[0]
```

```python
# Get list of object columns
obj_cols = test_data.select_dtypes(include=['object']).columns.tolist()

# Loop through object columns and factorize
for col in obj_cols:
    test_data[col] = pd.factorize(test_data[col])[0]
```

```python
#Matrix Factorization (MF) model using the Alternating Least Squares (ALS) algorithm:
from scipy.sparse import csr_matrix
from implicit.als import AlternatingLeastSquares
```

```python
# Create a sparse matrix of the training data
train_matrix = csr_matrix(train_data.values)

# Define the MF model
mf_model = AlternatingLeastSquares(factors=10, regularization=0.01)
```

```python
# Fit the model to the training data
mf_model.fit(train_matrix)
```

```python
test_matrix = csr_matrix(test_data.values)
item_factors = mf_model.item_factors
user_factors = mf_model.user_factors
test_predictions = item_factors.dot(user_factors.T)
```

```python
# Get the user index
user_index = train_data.index.get_loc(1)

# Get the user factors from the trained model
user_factors = mf_model.user_factors[user_index]

# Compute the item factors and their scores
item_factors = mf_model.item_factors
scores = item_factors.dot(user_factors)
```

```python
# Get the top recommendations
top_indices = np.argsort(scores)[::-1][:5]
recommended_items = test_data.index[top_indices]
```

```python
# Print the details of recommended laptops
for item_index in recommended_items:
    laptop_details = test_data1.loc[item_index]
    print(laptop_details)
    print('------------------------------------------------------------------------------')    
```
